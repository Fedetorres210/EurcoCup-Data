```python
import pandas as pd
from pymongo import MongoClient
from dotenv import load_dotenv
import os
import json
```

```python
load_dotenv()
url = os.getenv('URL')
```

```python
df = pd.read_csv("eurocup_2020_results.csv")
df
#Arreglar Date, Stage,pens,events, lineups.
```

```python
df = df.drop(columns=["events_list","lineup_home","lineup_away"])
df = df.drop(columns= ["duels_won_away","duels_won_home"])
df
```

```python
fecha = pd.to_datetime(df["date"])
date = fecha.dt.strftime("%d/%m/%y")
df["date"] = date
df
```

```python
df['team_name_home'].unique()
```

```python
lst = []
for elem in df['team_name_home']:
    lst.append(elem.replace(" ", ""))
    
```

```python
df['team_name_home'] = lst
```

```python
lst_1 = []
for elem in df['team_name_away']:
    lst_1.append(elem.replace(" ", ""))
df['team_name_away'] = lst_1
```

```python
df['stage'].unique()
```

```python
stages = []
for elem in df['stage']:
    stages.append(elem.replace(" ", ""))
```

```python
df['stage'] = stages
df
```

```python
data = df.to_dict(orient = "records")
```

```python
conn = MongoClient(url)
```

```python
db = conn.databaseEuro2020
```

```python
collection = db.EurocupData
```

```python
collection.insert_many(data)
```
